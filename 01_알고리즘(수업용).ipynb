# 하노이의 탑
### 입력: 옮기려는 원반의 갯수 n
###      옮길 원반이 현재 있는 출발점 기둥 from_pos
###      원반을 옮길 도착점 기둥 to_pos
###      옮기는 과정에서 사용할 보조 기둥 aux_pos
### 출력: 원반을 옮기는 순서 

 n층짜리 하노이탑을 옯기려면 원반을 몇번 움직여야할까?  
 2 <sup>n</sup>-1번으로 n이 커지면 -1은 의미가 없으므로 O(2<sup>n</sup>)입니다.  
 이는 기하급수적으로 증가하는 것으로 원반하나를 옮기는데 1초가 걸린다고 할 때  
 30층짜리 하노이탑을 옯기려면 무려 34시간을 먹지도 쉬지도 않고 원반만 옮겨야합니다.  

 #계산 복잡도
 ### o(1) : n과 무관하게 일정한 시간이 걸림
 ### o(n) : n과 비례하여 시간이 증가함
 ### o(n<sup>2</sup>) : n의 제곱에 비례하여 시간이 증가함
 ### o(2<sup>n</sup>) : 2의 n제곱에 비례하여 시간이 증가함
